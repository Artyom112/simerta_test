```python
data = pd.read_excel('data/Массив.xlsx')
```

```python
survey = data.iloc[:, -14:-1]
```

Видим желание использовать бесплатные дороги, если только не 100 рублей. Люди готовы заплатить 100 рублей за уменьшение времени в пути на 20 минут. Отчетливо прослеживатся желание использовать общественный транспорт.

```python
survey.iloc[:, 2].value_counts()
```

```python
plt.figure(figsize=(15, 40))
plt.tight_layout()
for idx in range(13):
    df = survey.iloc[:, idx]
    df = df[~df.isna()]
    
    plt.subplot(7, 2, idx+1)
    sns.countplot(data, y=survey.columns[idx], palette = "Set1")
    plt.xticks(fontsize=12, rotation=90)
    plt.xlabel('')
    plt.ylabel('')
    
plt.subplots_adjust(wspace=1.5)
plt.show()
```

```python
def find_time(hours, mins):
    if np.isnan(hours) and not np.isnan(mins):
        return mins
    elif not np.isnan(hours) and not np.isnan(mins):
        return hours * 60 + mins
    elif not np.isnan(hours) and np.isnan(mins):
        return hours * 60
    else:
        return None
```

```python
data['depart_time'] = data[[time_cols[0], time_cols[2]]].apply(lambda x: find_time(x[0], x[1]), axis=1)
```

```python
data['arrive_time'] = data[[time_cols[1], time_cols[3]]].apply(lambda x: find_time(x[0], x[1]), axis=1)
```

```python
data['travel_time'] = data[['arrive_time', 'depart_time']].apply(lambda x: x[0] - x[1], axis=1)
```
